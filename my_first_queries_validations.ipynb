```python
import pandas as pd
```

```python
data = pd.read_csv('AppleStore.csv')
data.sample(5)
```

```python
data[['prime_genre','track_name']].head()
```

```python
data[['rating_count_tot','prime_genre']].groupby('prime_genre').sum().sort_values('rating_count_tot',ascending=False).head()
```

```python
data[['track_name','prime_genre']].groupby('prime_genre').count().sort_values('track_name',ascending=False).head()
```

```python
data[['rating_count_tot','user_rating','track_name','prime_genre']].groupby(['track_name','prime_genre']).agg({'user_rating':'mean','rating_count_tot':'sum'}).sort_values('rating_count_tot',ascending=False).head(10)
```

```python
data[['track_name','user_rating','rating_count_tot','prime_genre','price']].groupby(['track_name','prime_genre']).agg({'user_rating':'mean','rating_count_tot':'sum','price':'mean'}).sort_values('price',ascending=False).head(10)
```
