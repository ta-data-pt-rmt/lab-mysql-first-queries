```python
import getpass
import mysql.connector
#saved_password = getpass.getpass()
# great to get a save the password into a variable to not reveal it
```

```python
cnx = mysql.connector.connect(user= 'root',
                       password= 'password',
                       host = 'localhost',
                       database='sakila',
                       auth_plugin='mysql_native_password')
```

```python
cnx.is_connected()
```

```python
if cnx.is_connected():
    print('Connection open')
    # do stuff you need to the database
else:
    print('Connection unsuccessful')
```

```python
cursor = cnx.cursor()
```

```python
query = ("SELECT * FROM sakila.actor;")
```

```python
cursor.execute(query)
```

```python
results = cursor.fetchall()
results
```

```python
for element in results:
    print('full name: '+element[1]+' '+element[2])
```

```python
import pandas as pd

actors_df = pd.DataFrame(results)
actors_df["full name"] = actors_df[1] + " " +  actors_df[2] 
actors_df
```

```python
query = ("""DROP DATABASE labs""")
cursor.execute(query)
```

```python
query = ("""CREATE DATABASE labs""")
cursor.execute(query)
```

```python
query = ("""CREATE TABLE IF NOT EXISTS
labs.demo(
student_id INT PRIMARY KEY,
student_name VARCHAR(52),
email VARCHAR(32) DEFAULT "john_doe@gmail.com",
coolness INT NOT NULL)""")
```

```python
cursor.execute(query)
```

```python
for index, firstname, lastname in zip(actors_df[0],actors_df[1],actors_df[2]):
    query = """INSERT INTO labs.demo(student_id,student_name,coolness) VALUES ("""+str(index)+""",'"""+firstname+" "+lastname+"""',10)"""
    #print(query)
    cursor.execute(query)
```

```python
cnx.commit()
# clear the cursor
cursor.close()

# closes the door between python and SQL
cnx.close()
```
